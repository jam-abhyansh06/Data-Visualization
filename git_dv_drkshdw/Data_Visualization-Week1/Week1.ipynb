```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)

df=pd.read_csv('googleplaystore1.csv')
```

```python
df.head()
```

```python
df.drop_duplicates(subset='App',keep='first',inplace=True)
```

```python
df.columns
```

```python
df.drop(['Last Updated','Current Ver','Android Ver'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Installs']=df['Installs'].map(lambda x: x.rstrip('+').lstrip())
df['Installs']=df['Installs'].str.replace(',','')

df['Price']=df['Price'].map(lambda x: x.lstrip('$').rstrip())
```

```python
df['Installs'].head()
```

```python
df['Price'].unique()
```

```python
df['Installs']=pd.to_numeric(df['Installs'])
df['Installs'].dtype
```

```python
df['Price']=pd.to_numeric(df['Price'])
df['Price'].dtype
```

```python
df.dtypes
```

```python
df['Content Rating'].unique()
```

```python
df['Type'].unique()
```

```python
fig1=px.pie(df, values='Installs',names='Category',title="Category-wise Download")
fig1.show()
```

```python
fig2=px.histogram(df,x='Content Rating',title='Content Rating')
fig2.show()
```

```python
fig3=px.histogram(df,x='Rating',title='Overall Rating')
fig3.show()
```

```python
game_df=df[df['Category']=='GAME']
game_df.head()
```

```python
fig4=px.pie(game_df, values='Installs',names='Genres',title="Genres-wise Game Downloads")
fig4.show()
```

```python
comm_df=df[df['Category']=='COMMUNICATION']

comm_df.sort_values('Installs',axis=0,ascending = False,inplace=True)

comm_df.head()
```

```python
fig5=px.bar(comm_df, x='App',y='Installs',title='Most popular Communication Application')
fig5.show()
```
